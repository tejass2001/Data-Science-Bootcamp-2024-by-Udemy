```python
import pandas as pd
```

```python
raw_csv_data = pd.read_csv('files/absenteeism_data.csv')
raw_csv_data
```

```python
df = raw_csv_data.copy()
df.info()
```

```python
# Dropping columns
df = df.drop(['ID'], axis=1)
```

```python
df['Reason for Absence'].unique()
```

```python
# Converting to dummy for "Reason for Absence"
reason_columns = pd.get_dummies(df['Reason for Absence'], drop_first=True)
reason_columns
```

```python
# Drop 'Reason for Absence' and add dummy var in it's place
df = df.drop(['Reason for Absence'], axis=1)
# Group the dummy variables
# Group 1-14, 15-17, 18-21, 22-28
reason_type_1 = reason_columns.loc[:, 1:14].max(axis=1)
reason_type_2 = reason_columns.loc[:, 15:17].max(axis=1)
reason_type_3 = reason_columns.loc[:, 18:21].max(axis=1)
reason_type_4 = reason_columns.loc[:, 22:].max(axis=1)

df = pd.concat([df, reason_type_1, reason_type_2, reason_type_3, reason_type_4], axis=1)
df
```

```python
df.columns.values
```

```python
column_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason 1', 'Reason 2', 'Reason 3', 'Reason 4']
df.columns = column_names
df.head()
```

```python
# Reorder columns
column_names_reordered = ['Reason 1', 'Reason 2', 'Reason 3', 'Reason 4', 
                          'Date', 'Transportation Expense', 'Distance to Work', 'Age',
                          'Daily Work Load Average', 'Body Mass Index', 'Education',
                          'Children', 'Pets', 'Absenteeism Time in Hours']
df = df[column_names_reordered]
df.head()
```

### Creating a checkpoint

```python
df_reason_mod = df.copy()
```

### Manipulating 'Date' column

```python
df_reason_mod['Date'] = pd.to_datetime(df_reason_mod['Date'], format= '%d/%m/%Y')
df_reason_mod.info()
```

### Extract month value

```python
list_months = []
for i in range(df_reason_mod.shape[0]):
    list_months.append(df_reason_mod['Date'][i].month)
```

```python
df_reason_mod['Month Value'] = list_months
```

### Extract day value

```python
def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
df_reason_mod['Day of the week'] = df_reason_mod['Date'].apply(date_to_weekday)
df_reason_mod.head()
```

```python
### Drop 'Date' column and reorder
df_reason_date_mod = df_reason_mod.copy()
df_reason_date_mod = df_reason_date_mod.drop(['Date'], axis=1)
df_reason_date_mod.columns.values
```

```python
column_order = ['Reason 1', 'Reason 2', 'Reason 3', 'Reason 4',
                'Month Value', 'Day of the week', 'Transportation Expense',
                'Distance to Work', 'Age','Daily Work Load Average', 
                'Body Mass Index', 'Education', 'Children', 'Pets', 'Absenteeism Time in Hours']
df_reason_date_mod = df_reason_date_mod[column_order]
df_reason_date_mod.head()
```

### Transform "Education" column

```python
df_reason_date_mod['Education'].unique()
```

```python
df_reason_date_mod['Education'].value_counts()
```

```python
# Combine 3,2,4 into single value
df_reason_date_mod['Education'] = df_reason_date_mod['Education'].map({1:0, 2:1, 3:1, 4:1})
```

```python
df_reason_date_mod['Education'].value_counts()
```

```python
# Checkpoint
df_preprocessed = df_reason_date_mod.copy()
```

### Exporting csv file

```python
df_preprocessed.to_csv('files/absenteeism_preprocessed.csv', index=False)
```
