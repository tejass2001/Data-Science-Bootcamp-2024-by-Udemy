### Multicollinearity

```python
data_cleaned.columns.values
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = data_cleaned[['Mileage', 'Year', 'EngineV']]
vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif['features'] = variables.columns
```

```python
vif
```

```python
data_no_multicollinearity = data_cleaned.drop(['Year'], axis=1)
```

```python
data_no_multicollinearity.describe(include='all')
```

### Creating dummmy variable

```python
 data_with_dummies = pd.get_dummies(data_no_multicollinearity, drop_first=True)
```

```python
data_with_dummies.head()
```

### Reorder the columns

```python
data_with_dummies.columns.values
```

```python
cols = ['log_price', 'Mileage', 'EngineV', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
```

```python
data_preprocessed = data_with_dummies[cols]
```

### VIF Check

```python
# Calculate the variance inflation factors for all variables contained in data_preprocessed. Anything strange?
variables = data_preprocessed
vif = pd.DataFrame()
vif["VIF"] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif["features"] = variables.columns
vif
```

Obviously, 'log_price' has a very high VIF. This implies it is most definitely linearly correlated with all the other variables. And this is no surprise! We are using a linear regression to determine 'log_price' given values of the independent variables! This is exactly what we expect - a linear relationship!

However, to actually assess multicollinearity for the predictors, we have to drop 'log_price'. The multicollinearity assumption refers only to the idea that the independent variables shoud not be collinear.

```python
# dropping the log_price and then calculating the VIF
variables = data_preprocessed
variables = variables.drop(['log_price'], axis=1)
vif = pd.DataFrame()
vif["VIF"] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif["features"] = variables.columns
vif
```

### Target and inputs

```python
targets = data_preprocessed['log_price']
inputs = data_preprocessed.drop(['log_price'], axis=1)
```

### Scaling the data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(inputs)

inputs_scaled = scaler.transform(inputs)
```

### Train test split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(inputs_scaled, targets, test_size=0.2, random_state = 42)
```

### Creating the regression

```python
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train, y_hat)
plt.xlabel("Targets (y_train)")
plt.ylabel("Predictions (y_hat)")
plt.show()
```

```python
# Residual plot
sns.distplot(y_train - y_hat)
```

```python
reg.score(x_train, y_train)
```

### Bias and coeff

```python
reg_summary = pd.DataFrame(inputs.columns.values, columns=['Features'])
reg_summary['Weights'] = reg.coef_
reg_summary
```

For dummy variables, Audi was dropped for brand. Hence all others are calculated with Audi as the baseline. 
Positive weight of brand_x means its expensive than Audi
Negative weight of brand_x means its cheaper than Audi

### Testing 

```python
y_hat_test = reg.predict(x_test)
```

```python
plt.scatter(y_test, y_hat_test, alpha=0.3)
plt.xlabel("Targets (y_test)")
plt.ylabel("Predictions (y_hat_test)")
plt.show()
```

```python
df_pf = pd.DataFrame(np.exp(y_hat_test), columns=['Predictions'])
df_pf.head()
```

```python
df_pf['Targets'] = np.exp(y_test)
df_pf.head()
```

```python
y_test
```

```python
# Need to reset the indices of y
y_test = y_test.reset_index(drop=True)
y_test.head()
```

```python
df_pf['Targets'] = np.exp(y_test)
df_pf.head()
```

```python
df_pf['Residual'] = df_pf['Targets'] - df_pf['Predictions']
```

```python
df_pf['Differences %'] = np.absolute(df_pf['Residual']/df_pf['Targets']*100)
df_pf
```
