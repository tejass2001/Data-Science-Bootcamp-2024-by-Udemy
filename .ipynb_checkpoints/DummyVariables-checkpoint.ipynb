```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
raw_data = pd.read_csv('files/1.03. Dummies.csv')
```

```python
raw_data
```

```python
data = raw_data.copy()
```

```python
data['Attendance'] = data['Attendance'].map({'Yes':1, 'No': 0})
```

```python
data
```

```python
data.describe()
```

```python
#Regression
y = data['GPA']
x1 = data[['SAT', 'Attendance']]
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
```

```python
results.summary()
```

```python
# Linear Equation:
# GPA = 0.0014*SAT + 0.2226*[0|1 (Attendance)] + 0.6439
```

```python
plt.scatter(data['SAT'], y)
yhat_no = 0.6439 + 0.0014*data['SAT']
yhat_yes = 0.8665 + 0.0014*data['SAT']
fig = plt.plot(data['SAT'], yhat_no, lw=2, c='#006837')
fig = plt.plot(data['SAT'], yhat_yes, lw=2, c='#a50037')
plt.xlabel('SAT')
plt.ylabel('GPA')
plt.show()
```

```python
plt.scatter(data['SAT'], y, c=data['Attendance'], cmap='RdYlGn_r')
yhat_no = 0.6439 + 0.0014*data['SAT']
yhat_yes = 0.8665 + 0.0014*data['SAT']
yhat = 0.0017*data['SAT'] + 0.275
fig = plt.plot(data['SAT'], yhat_no, lw=2, c='#006837')
fig = plt.plot(data['SAT'], yhat_yes, lw=2, c='#a50037')
fig = plt.plot(data['SAT'], yhat, lw=2, c='#4c72b0')
plt.xlabel('SAT')
plt.ylabel('GPA')
plt.show()
```

```python
# PREDICTING
# Bob got 1700 in SAT, Did not attend
# Alice got 1670 in SAT, Attended
x
```

```python
new_data = pd.DataFrame({'const': 1, 'SAT': [1700, 1670], 'Attendance': [0, 1]})
new_data = new_data.rename(index={0:'Bob', 1:'Alice'})
new_data
```

```python
predictions = results.predict(new_data)
predictions
```

```python
predictions_df = pd.DataFrame({'Predictions': predictions})
joined = new_data.join(predictions_df)
joined.rename(index={0: 'Bob', 1: 'Alice'})
```
